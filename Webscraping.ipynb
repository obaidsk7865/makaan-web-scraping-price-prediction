```python
import pandas as pd
```

```python
base_url='https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page='
for i in range(1,11):
    print(base_url+str(i))
```

```python
import time
from IPython.display import clear_output
base_url='https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page='
Builder=[]
no_bhk=[]
locality=[]
price=[]
area=[]
cons_stats=[]
for i in range(1,101):
    url = base_url+str(i)
    print(f'Scraping info from page {i}')
    clear_output(wait=True)
    time.sleep(3)
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html')
    
    #owner info
    own=soup.find_all('span', attrs={'class' : 'seller-type'})
   
    for o in own:
        Builder.append(o.text)
   
    
    #no_bhk
    bh=soup.find_all('a',attrs={'class' : 'typelink'})
    
    for b in bh:
        no_bhk.append(b.span.text)
    
    
    #locality
    loc=soup.find_all('span',attrs={'itemprop' : 'addressLocality'})
   
    for l in loc:
        locality.append(l.text)
    
    
    
    #price
    
    import re
    prc=soup.find_all('div',attrs={'data-type' : 'price-link'})   
    for i in prc:
        units = re.sub('[^a-zA-Z]', '', i.text)
        p = i.text
        if units=='Cr':
            p = p.replace('Cr','')
            p = float(p)*10000000
        elif units=='L':
            p = p.replace('L','')
            p = float(p)*100000
        price.append(int(p))

    
    
    #area
    
    ar=soup.find_all('td',attrs={'class' : 'size'})
    
    for a in ar:
        area.append(a.text)
    
    
    #cons-stat
    stat=soup.find_all('td',attrs={'class' : 'val'})
   
    for s in stat:
        cons_stats.append(s.text)
    
    
    
```

```python
print(len(price))
```

```python
data=pd.DataFrame({'no_bhk':no_bhk,'price':price,'area':area,'cons_stats':cons_stats})
data
```

```python
data.to_csv('realestate.csv')
```
